### About Dataset
**Description**

Pulsars are a rare type of Neutron star that produce radio emission detectable here on Earth. They are of considerable scientific interest as probes of space-time, the inter-stellar medium, and states of matter. Machine learning tools are now being used to automatically label pulsar candidates to facilitate rapid analysis. Classification systems in particular are being widely adopted,which treat the candidate data sets as binary classification problems.

Credit goes to Pavan Raj (https://www.kaggle.com/pavanraj159) from where the dataset has been collected. For the purpose of creating a challenge, certain modifications have been done to the dataset.

Original dataset can be acquired from the link Predicting a Pulsar Star (https://www.kaggle.com/pavanraj159/predicting-a-pulsar-star)
Attribute Information:

Each candidate is described by 8 continuous variables, and a single class variable. The first four are simple statistics obtained from the integrated pulse profile (folded profile). This is an array of continuous variables that describe a longitude-resolved version of the signal that has been averaged in both time and frequency . The remaining four variables are similarly obtained from the DM-SNR curve . These are summarised below:

1. Mean of the integrated profile.
2. Standard deviation of the integrated profile.
3. Excess kurtosis of the integrated profile.
4. Skewness of the integrated profile.
5. Mean of the DM-SNR curve.
6. Standard deviation of the DM-SNR curve.
7. Excess kurtosis of the DM-SNR curve.
8. Skewness of the DM-SNR curve.
9. Class

HTRU 2 Summary

1,639 positive examples.
16,259 negative examples.

## Train Test Split

```python
# perform train-test-split

#read in again to remove previous changes
new_columns = ['mean_profile', 'std_profile', 'kurtosis_profile', 'skewness_profile',
               'mean_DM_SNR', 'std_DM_SNR', 'kurtosis_DM_SNR', 'skewness_DM_SNR', 'target']

df = pd.read_csv('archive/pulsar_data_train.csv')
df.columns = new_columns
```

```python
# for baseline drop all nans
df = df.dropna()
```

```python
# define features and target
target = df['target']
features = df.drop(columns='target')

features_train, features_test, target_train, target_test = train_test_split(features, 
                                                                            target, 
                                                                            random_state=42,
                                                                            test_size=0.2)
```

```python
print(features_train.shape, features_test.shape)
```

## Basline Model

```python
from imblearn.over_sampling import SMOTE, RandomOverSampler
from imblearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import regularizers
from tensorflow.keras.layers import Dropout
```

```python
scaler = StandardScaler()

# apply preprocessing
features_train_std = scaler.fit_transform(features_train)
features_test_std = scaler.transform(features_test)

# Apply SMOTE 
smote = SMOTE()
features_train_resampled, target_train_resampled = smote.fit_resample(features_train_std, target_train)
```

```python

# define model
model_ann_early = Sequential()

# define hidden layers
hidden_1 = Dense(units=50, activation='relu', input_dim=features_train.shape[1])
hidden_2 = Dense(units=50, activation='relu')
hidden_3 = Dense(units=50, activation='relu')
hidden_4 = Dense(units=50, activation='relu')
hidden_5 = Dense(units=50, activation='relu')

# define the dropout layers
dropout_1 = Dropout(rate=0.3)
dropout_2 = Dropout(rate=0.3)
dropout_3 = Dropout(rate=0.3)
dropout_4 = Dropout(rate=0.3)
dropout_5 = Dropout(rate=0.3)

# define output layer
output_layer  = Dense(units=1, activation='sigmoid')

# add  hidden layers 
model_ann_early.add(hidden_1)
model_ann_early.add(dropout_1)
model_ann_early.add(hidden_2)
model_ann_early.add(dropout_2)
model_ann_early.add(hidden_3)
model_ann_early.add(dropout_3)
#model_ann_early.add(hidden_4)
#model_ann_early.add(dropout_4)
#model_ann_early.add(hidden_5)
#model_ann_early.add(dropout_5)
#model_ann_early.add(hidden_4)
#model_ann_early.add(hidden_5)



# add output layer
model_ann_early.add(output_layer)

#-----------------------------------------------------------------------------------------

import tensorflow.keras.backend as K
from tensorflow.keras.metrics import Recall, Precision

def f1_score_ann(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))

    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())

    f1 = 2 * (precision * recall) / (precision + recall + K.epsilon())
    return f1
#-----------------------------------------------------------------------------------------
from tensorflow.keras.optimizers import Adam
adam_custom = Adam(learning_rate=0.00005)

#-----------------------------------------------------------------------------------------
    

model_ann_early.compile(optimizer='adam', loss='binary_crossentropy', metrics=[f1_score_ann])
#model_ann_early.compile(optimizer=adam_custom, loss='binary_crossentropy', metrics=[f1_score_ann])
```

```python
early_stop = EarlyStopping(monitor='f1_score_ann', min_delta=0.0001, patience=3)

# train model using early stopping
hist_ann = model_ann_early.fit(features_train_resampled,
                               target_train_resampled,
                               epochs=20, 
                               batch_size=128,
                               callbacks=[early_stop],
                               validation_data=(features_test_std, target_test))
```

```python
# define figure and axes
fig, axs = plt.subplots(ncols=2, figsize=(12,6))

# plot training & validation loss values
axs[0].plot(hist_ann.history['loss'])
axs[0].plot(hist_ann.history['val_loss'])
axs[0].set(title='Model loss', ylabel='Loss', xlabel='Epoch')
axs[0].legend(['train', 'val'])
axs[0].set_xticks(range(0,len(hist_ann.history['loss'])))
#axs[0].set_xtickslabels(range(1,len(hist_ann.history['loss'])))
#axs[0].set_xticks(range(10))
#axs[0].set_xticklabels(range(1, 11))

# plot training & validation accuracy values
axs[1].plot(hist_ann.history['f1_score_ann'])
axs[1].plot(hist_ann.history['val_f1_score_ann'])
axs[1].set(title='Model f1_score', ylabel='f1', xlabel='Epoch')
axs[1].legend(['train', 'val'], loc='upper left')
axs[1].set_xticks(range(0,len(hist_ann.history['loss'])));
#axs[1].set_xticks(range(10))
#axs[1].set_xticklabels(range(1, 11));
```
